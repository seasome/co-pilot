```python
import json

fname = "data.tsv"

template = {
    "objects": [{
        "properties": {
            "enaio:objectTypeId": {
                "value": "copilot"
            },
            "question": {
                "value": ""
            },
            "answer": {
                "value": ""
            },
            "type": {
                "value": ""
            },
            "misc": {
                "value": ""
            }
        }
        }]
    }
with open(fname) as f:
    for idx, line in enumerate(f):
        if idx == 0: continue
        query, metadata, label = line.strip().split("\t")
        with open("metafiles/{0}.json".format(idx),"w") as fw:
            template["objects"][0]["properties"]["question"]["value"] = query
            #template["objects"][0]["properties"]["answer"]["value"] 
            template["objects"][0]["properties"]["type"]["value"] = label
            template["objects"][0]["properties"]["misc"]["value"] = metadata
            json.dump(template,fw)
```
